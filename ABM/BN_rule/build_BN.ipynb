```python
from pomegranate import *
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import networkx
from pomegranate.utils import plot_networkx
import bnlearn
```

```python
survey_data_loc = 'C:/Users/pyangac/Documents/research/cabbi/ABM_pilot/BN_rule/'
survey_data = pd.read_csv(survey_data_loc+'BN_data_start1_adj.csv')
```

```python
survey_data
```

```python
edges = [('farm_size', 'imp_env'),
         ('info_use', 'imp_env'),
         ('info_use', 'max_fam'),
         ('info_use', 'lql'),
        ('peer_ec','max_fam'),
        ('com_sup','benefit'),
        ('risk','lql'),
        ('benefit','SC_Will'),
        ('concern','SC_Will'),
        ('imp_env','SC_Will'),
        ('max_fam','SC_Will'),
        ('lql','SC_Will'),
        ('SC_Contract','SC_Will'),
        ('SC_Contract','SC_Cost_share'),
        ('SC_Cost_share','SC_Will'),
        ('SC_Rev_mean','SC_Will'),
        ('SC_Rev_range','SC_Will'),
        ('SC_Subsidy','SC_Will'),
        ('SC_Will','SC_Ratio'),
        ('SC_Contract','SC_Ratio'),
        ('SC_Cost_share','SC_Ratio'),
        ('SC_Rev_mean','SC_Ratio'),
        ('SC_Rev_range','SC_Ratio'),
        ('SC_Subsidy','SC_Ratio')]
# Make the actual Bayesian DAG
model = bnlearn.make_DAG(edges)
```

```python
model = bnlearn.parameter_learning.fit(model, survey_data)
```

```python
q1 = bnlearn.inference.fit(model, variables=['SC_Will'], evidence={'farm_size':1})
```

```python
q1 = bnlearn.inference.fit(model, variables=['SC_Ratio'], evidence={'SC_Rev_mean':3})
```

```python
q1 = bnlearn.inference.fit(model, variables=['SC_Will'], evidence={'SC_Rev_mean':1,'farm_size':2})
```
